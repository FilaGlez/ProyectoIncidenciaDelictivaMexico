# Adding LAT and LON in Data

```python
## Adding to incidencia delictiva municipal
#Import both DF
import pandas as pd
import numpy as np

path_inegi = 'D:\\Downloads\\'
df_inegi = pd.read_excel(path_inegi + "AGEEML_2023101313538.xlsx", skiprows=3) # INEGI
path_mun = "D:\\Proyects\\IncidenciaDelictivaMexico\\Data\\"
df_mun = pd.read_csv(path_mun + "IDM_NM_ago23.csv", encoding='iso-8859-1') #incidencia delictiva municipal
```

```python
#Column of both dfs
print("Columnas Inegi: ",df_inegi.columns)
print("Columnas municipios: ",df_mun.columns)
```

```python
#Selecting columns

#INEGI
df_inegi_test = df_inegi[["NOM_ENT","NOM_MUN","NOM_LOC","LAT_DECIMAL","LON_DECIMAL"]] #INEGI
#MUNICIPIO
df_mun = df_mun.drop(columns=['Clave_Ent', 'Cve. Municipio']) #Dropping columns not needed
```

#### MUNICIPIO 

```python
#Eliminating space between words of Entidad and Municipio and adding them together to create a unique id for MUN
df_mun['Municipio'] = df_mun['Municipio'].str.strip()
df_mun['Municipio'] = df_mun['Municipio'].str.replace('Angel Albino Corzo', 'Ángel Albino Corzo') #Change value of specific row
df_mun['Municipio_'] = df_mun['Municipio'].str.replace(' ', '')
df_mun['Entidad'] = df_mun['Entidad'].str.strip()
df_mun['Entidad_'] = df_mun['Entidad'].str.replace(' ', '')
df_mun['Concatenated_Column'] = df_mun['Entidad_'] + df_mun['Municipio_']
df_mun = df_mun.drop(columns=['Municipio_', 'Entidad_']) #Dropping not needed columns
df_mun.columns
```

#### INEGI

```python
# Eliminating space between words of Entidad and Municipio and adding them together to create a unique id for INEGI
df_inegi_test['NOM_MUN'] = df_inegi_test['NOM_MUN'].str.replace('Angel Albino Corzo', 'Ángel Albino Corzo')#Change value of specific row
df_inegi_test['NOM_MUN'] = df_inegi_test['NOM_MUN'].str.strip()
df_inegi_test['NOM_MUN_'] = df_inegi_test['NOM_MUN'].str.replace(' ', '')
df_inegi_test['NOM_ENT'] = df_inegi_test['NOM_ENT'].str.strip()
df_inegi_test['NOM_ENT_'] = df_inegi_test['NOM_ENT'].str.replace(' ', '')
df_inegi_test['Concatenated_Column'] = df_inegi_test['NOM_ENT_'] + df_inegi_test['NOM_MUN_']
df_inegi_test = df_inegi_test.drop(columns=['NOM_MUN_', 'NOM_ENT_']) #Dropping not needed columns
df_inegi_test.columns
```

```python
#Getting distinct values of id columns

distinct_mun_concat = df_mun['Concatenated_Column'].drop_duplicates() #MUN
distinct_inegi = df_inegi_test['Concatenated_Column'].drop_duplicates() #INEGI
```

```python
#Import states with capitals added
states = pd.read_excel(path_mun + "Edos.xlsx")
#Taking the space between words for capitales
states['Capitales'] = states['Capitales'].str.strip()
states['Capitales_'] = states['Capitales'].str.replace(' ', '')
#Taking the space between words for Entidades
states['Entidad'] = states['Entidad'].str.strip()
states['Entidad_'] = states['Entidad'].str.replace(' ', '')
states['Concatenated_Column'] = states['Entidad_'] + states['Capitales_']
states.tail()
```

```python
#INEGI table - Obtaining the id for state and capital

#df_inegi_test.tail()
#Taking the space between words for ENT
df_inegi_test['NOM_ENT'] = df_inegi_test['NOM_ENT'].str.strip()
df_inegi_test['NOM_ENT_'] = df_inegi_test['NOM_ENT'].str.replace(' ', '')
#Taking the space between words for LOC
df_inegi_test['NOM_LOC'] = df_inegi_test['NOM_LOC'].str.strip()
df_inegi_test['NOM_LOC_'] = df_inegi_test['NOM_LOC'].str.replace(' ', '')
#Values not present, replacing values to correspond to the capital
df_inegi_test['NOM_LOC_'] = df_inegi_test['NOM_LOC_'].str.replace('Xalapa-Enríquez', 'XalapaEnríquez') 
df_inegi_test['NOM_LOC_'] = df_inegi_test['NOM_LOC_'].str.replace('SanFranciscodeCampeche', 'Campeche')
df_inegi_test['NOM_LOC_'] = df_inegi_test['NOM_LOC_'].str.replace('Cuauhtémoc', 'CiudaddeMéxico')
#Creating new column of ENT and LOC
df_inegi_test['Concatenated_Column'] = df_inegi_test['NOM_ENT_'] + df_inegi_test['NOM_LOC_']

df_inegi_test.tail()
```

```python
#FINDING MISSING VALUES
#Getting distinct values of id columns

distinct_edo_concat = states['Concatenated_Column'].drop_duplicates() #EDO
distinct_inegi = df_inegi_test['Concatenated_Column'].drop_duplicates() #INEGI
print(distinct_edo_concat)
```

```python
# Find values in inegi but not in mun
in_inegi_not_in_edo = set(distinct_inegi) - set(distinct_edo_concat)

# Find values in mun but not in inegi
in_edo_not_in_inegi = set(distinct_edo_concat) - set(distinct_inegi)

#print("Values in inegi but not in mun:", in_inegi_not_in_edo)
print("Values in edo but not in inegi:", in_edo_not_in_inegi)
```

```python
#Adding missing columns of LAT and LOG to EDO
df_inegi_merge = df_inegi_test[["Concatenated_Column","LAT_DECIMAL","LON_DECIMAL"]]
inegi = pd.DataFrame(df_inegi_merge)
states = pd.DataFrame(states)
#states.tail()
# Merge the 'inegi' and 'states' DataFrames based on 'Concatenated_Column'

# Merge the 'inegi' and 'states' DataFrames based on 'Concatenated_Column'
result = inegi.merge(states, on='Concatenated_Column')

# Display the result
print(result)
```
